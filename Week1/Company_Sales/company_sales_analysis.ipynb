# Analyzing a Company Sales Data with Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales_data = pd.read_csv('./company_sales_data.csv')
sales_data
```

### Task (a): Generating a line plot with the total_profit column

```python
plt.plot('month_number', 'total_profit', 'bo-', data=sales_data)

plt.xlabel('Months')
plt.ylabel('Total Profit')
plt.ylim(150000, 450000)

plt.show()
```

#### Task (b): Displaying Bathing_Soap and Fashwash on Subplot

```python
# Bathing_soup subplot:
plt.subplot(2, 1, 1)
plt.plot('month_number', 'bathingsoap', data=sales_data)
plt.ylabel('bathingsoup')

# Fashwash subplot:
plt.subplot(2, 1, 2)
plt.plot('month_number', 'facewash', data=sales_data)
plt.ylabel('facewash')

plt.xlabel('month_number') # setting the horizontal label
plt.suptitle('Subplots for BathingSoup and Facewash columns')   # setting title for the subplot
plt.show()
```
